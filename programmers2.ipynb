https://school.programmers.co.kr/learn/courses/30/lessons/299310

```python
SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR,
MAX(SIZE_OF_COLONY) OVER (PARTITION BY YEAR(DIFFERENTIATION_DATE)) - SIZE_OF_COLONY AS YEAR_DEV,
ID
FROM ECOLI_DATA
ORDER BY YEAR, YEAR_DEV ASC;
```

##신고 결과 받기

https://school.programmers.co.kr/learn/courses/30/lessons/92334

```python
from collections import defaultdict


def solution(id_list, report, k):

    users = {id: {"target": set(), "banned_cnt": 0} for id in id_list}
    report_cnt = defaultdict(int)

    for item in report:
        reporter, target = item.split()

        if target not in users[reporter]["target"]:
            users[reporter]["target"].add(target)
            report_cnt[target] += 1

    for user, cnt in report_cnt.items():
        if cnt >= k:

            for reporter, target in users.items():
                if user in target["target"]:
                    users[reporter]["banned_cnt"] += 1


    answer = [user["banned_cnt"] for user in users.values()]
    return answer
```

##성격 유형 검사하기

https://school.programmers.co.kr/learn/courses/30/lessons/118666

```python
from collections import defaultdict


def solution(survey, choices):
    answer = []

    tables = ["RT", "CF", "JM", "AN"]
    scores = {1: 3, 2: 2, 3: 1, 5: 1, 6: 2, 7: 3}
    types = defaultdict(int)

    for i in range(len(survey)):
        no, yes = survey[i][0], survey[i][1]

        if choices[i] in [1, 2, 3]:
            types[no] += scores[choices[i]]
        elif choices[i] in [5, 6, 7]:
            types[yes] += scores[choices[i]]

    for word in tables:
        if types[word[0]] > types[word[1]] or types[word[0]] == types[word[1]]:
            answer.append(word[0])
        else:
            answer.append(word[1])

    return "".join(answer)
```

##개인정보 수집 유효기간


https://school.programmers.co.kr/learn/courses/30/lessons/150370

```python
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta


def get_date(date):
    return datetime.strptime(date, "%Y.%m.%d")


def solution(today, terms, privacies):
    today = get_date(today)
    answer = []
    tm = {}

    for t in terms:
        term, mon = t.split(" ")
        tm[term] = int(mon)

    for i, p in enumerate(privacies):
        date, term = p.split(" ")
        time = tm[term]
        start_date = get_date(date)
        end_date = start_date + relativedelta(months=time)

        if end_date <= today:
            answer.append(i+1)

    return answer
```

##[PCCE 기출문제] 8번 / 창고 정리

https://school.programmers.co.kr/learn/courses/30/lessons/250126

```python
def solution(storage, num):
    clean_storage = []
    clean_num = []
    for i in range(len(storage)):
        if storage[i] in clean_storage:
            pos = clean_storage.index(storage[i])
            clean_num[pos] += num[i]
        else:
            clean_storage.append(storage[i])
            clean_num.append(num[i])

    # 아래 코드에는 틀린 부분이 없습니다.

    max_num = max(clean_num)
    answer = clean_storage[clean_num.index(max_num)]
    return answer
```

##[PCCE 기출문제] 10번 / 데이터 분석

https://school.programmers.co.kr/learn/courses/30/lessons/250121

```python
from functools import cmp_to_key


def solution(data, ext, val_ext, sort_by):
    cols = {"code": 0, "date": 1, "maximum": 2, "remain": 3}

    def cmp(x, y):
        return x[cols[sort_by]] - y[cols[sort_by]]

    data = [row for row in data if row[cols[ext]] <= val_ext]

    answer = sorted(data, key=cmp_to_key(cmp))
    return answer
```

##[PCCE 기출문제] 9번 / 이웃한 칸

https://school.programmers.co.kr/learn/courses/30/lessons/250125

```python
def solution(board, h, w):
    answer = 0
    color = board[h][w]
    directions = [[0, -1], [0, 1], [-1, 0], [1, 0]]

    for dir_h, dir_w in directions:
        mov_h = h + dir_h
        mov_w = w + dir_w

        if (0 <= mov_h <= len(board)-1) and (0 <= mov_w <= len(board[0])-1):
            answer += int(color == board[mov_h][mov_w])

    return answer
```

##[PCCE 기출문제] 5번 / 산책

https://school.programmers.co.kr/learn/courses/30/lessons/250129?language=python3

```python
def solution(route):
    east = 0
    north = 0
    for i in route:
        if i == "N":
            north += 1
        elif i == "S" :
            north -= 1
        elif i == "E" :
            east += 1
        elif i == "W" :
            east -= 1

    return [east, north]
```

##[PCCP 기출문제] 1번 / 붕대 감기

https://school.programmers.co.kr/learn/courses/30/lessons/250137

```python
def solution(bandage, health, attacks):
    answer = 0
    t, x, y = bandage
    max_health = health
    success_heal = 0
    map_attacks = {attack[0]:attack[1] for attack in attacks}
    last_attack = attacks[-1][0]

    for time in range(1,last_attack+1):
        if time in map_attacks:
            health -= map_attacks[time]
            success_heal = 0
            if health <= 0:
                return -1
        else:
            success_heal += 1
            if success_heal == t:
                health = min(health+x+y, max_health)
                success_heal = 0
            else:
                health = min(health+x, max_health)

    answer = health
    return answer
```

##부모의 형질을 모두 가지는 대장균 찾기

https://school.programmers.co.kr/learn/courses/30/lessons/301647

```python
SELECT
    C.ID
    , C.GENOTYPE
    , P.GENOTYPE PARENT_GENOTYPE
FROM
    ECOLI_DATA C
    INNER JOIN ECOLI_DATA P
        ON C.PARENT_ID = P.ID
WHERE
    C.GENOTYPE & P.GENOTYPE = P.GENOTYPE
ORDER BY
    C.ID
```

##없어진 기록 찾기

https://school.programmers.co.kr/learn/courses/30/lessons/59042

```python
SELECT
O.ANIMAL_ID, O.NAME
FROM
ANIMAL_OUTS O
LEFT JOIN
ANIMAL_INS I
ON
O.ANIMAL_ID = I.ANIMAL_ID
WHERE
I.ANIMAL_ID IS NULL
ORDER BY
O.ANIMAL_ID
```
