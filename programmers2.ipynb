##연도별 대장균 크기의 편차 구하기

https://school.programmers.co.kr/learn/courses/30/lessons/299310

```python
SELECT YEAR(DIFFERENTIATION_DATE) AS YEAR,
MAX(SIZE_OF_COLONY) OVER (PARTITION BY YEAR(DIFFERENTIATION_DATE)) - SIZE_OF_COLONY AS YEAR_DEV,
ID
FROM ECOLI_DATA
ORDER BY YEAR, YEAR_DEV ASC;
```

##특정 형질을 가지는 대장균 찾기


https://school.programmers.co.kr/learn/courses/30/lessons/301646

```python
SELECT COUNT(1) COUNT
FROM ECOLI_DATA
WHERE (GENOTYPE & 2) != 2 AND ((GENOTYPE & 4) = 4 OR (GENOTYPE & 1) = 1)
```

##같은 숫자는 싫어

https://school.programmers.co.kr/learn/courses/30/lessons/12906

```python
def solution(arr):
    answer = [arr[0]]
    for i in range(1,len(arr)):
        if arr[i]!=answer[-1]:
            answer.append(arr[i])

    return answer
```

##완주하지 못한 선수


https://school.programmers.co.kr/learn/courses/30/lessons/42576

```python
def solution(participant, completion):
    answer = ''
    n=len(participant)
    dic1={i:0 for i in participant}
    dic2={i:0 for i in participant}

    for i in participant:
        dic1[i]+=1
    for i in completion:
        dic2[i]+=1

    for i in dic1:
        if dic1[i]-dic2[i]==1:
            return i
```

##신고 결과 받기

https://school.programmers.co.kr/learn/courses/30/lessons/92334

```python
from collections import defaultdict


def solution(id_list, report, k):

    users = {id: {"target": set(), "banned_cnt": 0} for id in id_list}
    report_cnt = defaultdict(int)

    for item in report:
        reporter, target = item.split()

        if target not in users[reporter]["target"]:
            users[reporter]["target"].add(target)
            report_cnt[target] += 1

    for user, cnt in report_cnt.items():
        if cnt >= k:

            for reporter, target in users.items():
                if user in target["target"]:
                    users[reporter]["banned_cnt"] += 1


    answer = [user["banned_cnt"] for user in users.values()]
    return answer
```
