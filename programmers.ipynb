##정수를 나선형으로 배치하기


https://school.programmers.co.kr/learn/courses/30/lessons/181832

```python
def solution(n):
    x, y = 0, 0
    direction = "right"

    L = [[0 for _ in range(n)] for __ in range(n)]


    for num in range(1, n * n + 1):


        L[x][y] = num


        if direction == "right":

            if y + 1 == n or L[x][y + 1] != 0:
                direction = "down"
                x += 1
            else:
                y += 1

        elif direction == "down":

            if x + 1 == n or L[x + 1][y] != 0:
                direction = "left"
                y -= 1
            else:
                x += 1

        elif direction == "left":

            if L[x][y - 1] != 0:
                direction = "top"
                x -= 1
            else:
                y -= 1

        elif direction == "top":

            if L[x - 1][y] != 0:
                direction = "right"
                y += 1
            else:
                x -= 1
    return L
```

```python
def solution(keyinput, board):

    x, y = 0, 0

    max_x, max_y = board[0] // 2, board[1] // 2


    for key in keyinput:

        if key == "up" and y < max_y:
            y += 1

        elif key == "down" and y > -max_y:
            y -= 1

        elif key == "left" and x > -max_x:
            x -= 1

        elif key == "right" and x < max_x:
            x += 1

    return [x, y]
```

##다항식 더하기


https://school.programmers.co.kr/learn/courses/30/lessons/120863

```python
def solution(polynomial):
    answer = ''
    digit_counter = []
    alpha_counter = []
    alnum_counter = []

    re_alnum = []

    poly_split = polynomial.split(' ')
    print(poly_split)

    for x in poly_split:
        if x.isdigit() == True:
            digit_counter.append(int(x))
        elif x.isalpha() == True:
            alpha_counter.append(x)
        elif x.isalnum() == True:
            alnum_counter.append(x)
    for alnum in alnum_counter:
        re_alnum.append(int(alnum.replace('x','')))

    digit_r = sum(digit_counter)
    alpha_r = len(alpha_counter)
    alnum_r = sum(re_alnum)
    x_result = alpha_r+alnum_r

    if digit_r == 0:
        if x_result == 1:
            answer = "x"
        else:
            answer = str(x_result)+"x"
    elif x_result == 0:
        answer = str(digit_r)
    elif digit_r != 0 and x_result != 0:
        if x_result == 1:
            answer = "x + "+str(digit_r)
        else:
            answer = str(x_result)+"x + "+str(digit_r)

    return answer
```

##평행


https://school.programmers.co.kr/learn/courses/30/lessons/120875

```python
def solution(dots):
    answer = 0
    [[x1, y1], [x2, y2], [x3, y3], [x4, y4]] = dots


    gradient_12 = abs((y2 - y1) / (x2 - x1))
    gradient_34 = abs((y4 - y3) / (x4 - x3))

    gradient_13 = abs((y3 - y1) / (x3 - x1))
    gradient_24 = abs((y4 - y2) / (x4 - x2))

    gradient_23 = abs((y3 - y2) / (x3 - x2))
    gradient_14 = abs((y4 - y1) / (x4 - x1))

    if gradient_12 == gradient_34 or gradient_23 == gradient_14 or gradient_13 == gradient_24:
        answer = 1


    return answer
```

```python
def solution(lines):
    line_list = []
    for line in lines:
        line_list += line
    mx, mn = max(line_list), min(line_list)

    x, y, z = lines
    overlap_line_count = 0
    for i in range(mn, mx+1):
        overlap_point_count = 0

        if x[0] <= i < x[1]:
            overlap_point_count += 1
        if y[0] <= i < y[1]:
            overlap_point_count += 1
        if z[0] <= i < z[1]:
            overlap_point_count += 1

        if overlap_point_count >= 2:
            overlap_line_count += 1


    return overlap_line_count
```

##유한소수 판별하기


https://school.programmers.co.kr/learn/courses/30/lessons/120878

```python
import math

def solution(a, b):
    if a==b: return 1
    c = b // math.gcd(a,b)
    while True:
        if c%2: break
        c = c//2
    while True:
        if c%5: break
        c = c//5

    return 1 if c==1 else 2
```

##대충 만든 자판


https://school.programmers.co.kr/learn/courses/30/lessons/160586

```python
def solution(keymap, targets):
    answer = []
    alp_key = dict()
    for row in keymap:
        for idx, alp in enumerate(row):

            if (alp in alp_key) and (idx+1) > alp_key[alp]:
                continue

            alp_key[alp] = idx + 1

    for row in targets:
        total_index = 0

        for alp in row:

            if alp in alp_key:
                total_index += alp_key[alp]

            else:
                total_index = -1
                break

        answer.append(total_index)
    return answer
```

##바탕화면 정리


https://school.programmers.co.kr/learn/courses/30/lessons/161990

```python
def solution(wallpaper):
    answer = []
    min_y, min_x = len(wallpaper), len(wallpaper[0])
    max_y, max_x = -1, -1


    for y_idx, x in enumerate(wallpaper):

        if "#" in x:

            if (y_idx < min_y):
                min_y = y_idx

            if (y_idx+1 > max_y):
                max_y = y_idx+1

            if (x.find("#") < min_x):
                min_x = x.find("#")

            if (x.rfind("#")+1 > max_x):
                max_x = x.rfind("#")+1

    answer = [min_y,min_x,max_y,max_x]
    return answer
```

##문자열 나누기


https://school.programmers.co.kr/learn/courses/30/lessons/140108

```python
def solution(s):
    answer = 0
    first_alp = ""
    s_count = 0
    d_count = 0

    for alp in s:

        if first_alp == "":
            first_alp=alp
            s_count=1
            continue

        if first_alp == alp:
            s_count+=1
        else:
            d_count+=1

        if s_count == d_count:
            first_alp = ""
            s_count=0
            d_count=0
            answer+=1

    if first_alp !="":
        answer +=1


    return answer
```

```python
def solution(number, limit, power):
    answer = 0

    for i in range(1, number+1):
        div_num=0

        for j in range(1,int(i**0.5)+1):
            if i % j == 0 :
                div_num += 2

        if i**0.5%1==0:
            div_num-=1

        if div_num<=limit:
            answer += div_num
        else:
            answer += power

    return answer
```

##숫자 짝꿍


https://school.programmers.co.kr/learn/courses/30/lessons/131128

```python
def solution(X, Y):
    answer = ''
    xy=set(X)&set(Y)

    if not xy:
        return '-1'

    elif len(xy) == 1 and '0' in xy:
        return '0'

    answer = [n*min(X.count(n),Y.count(n)) for n in xy]
    answer = ''.join(sorted(answer, reverse=True))
    return answer
```
